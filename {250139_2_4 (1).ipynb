## What are Functions in Programming?

In programming, a **function** is a block of organized, reusable code that performs a specific task. Functions are fundamental building blocks that allow you to break down a larger program into smaller, manageable, and modular pieces.

### Benefits of Using Functions:

1.  **Reusability**: Once a function is defined, it can be called multiple times from different parts of a program, avoiding redundant code and promoting the DRY (Don't Repeat Yourself) principle.
2.  **Modularity**: Functions help in organizing code into logical blocks, making the program structure clearer and easier to understand, debug, and maintain.
3.  **Readability**: By abstracting complex logic into named functions, the main program flow becomes more readable and easier to follow.
4.  **Easier Debugging**: When an issue arises, it's often easier to locate and fix problems within a small, focused function rather than sifting through a large block of code.

### Basic Syntax for Defining Functions in Python:

In Python, functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. Any input parameters or arguments are placed inside the parentheses. The function body is indented below the `def` line.

```python
def function_name(parameters):
    """Docstring: Explain what the function does."""
    # Function body (code to be executed)
    # ...
    return result # Optional: return a value
```

*   `def`: Keyword to indicate the start of a function definition.
*   `function_name`: A descriptive name for the function (follows Python's naming conventions).
*   `parameters`: Optional input values passed into the function (e.g., `arg1, arg2`).
*   `:`: Marks the end of the function header.
*   `Docstring`: An optional string literal used to document the function's purpose, arguments, and return values.
*   `return`: Optional keyword to send a value back to the caller of the function.

## Define Financial Indicator Functions

### Subtask:
Create six separate Python functions to calculate basic financial indicators: Simple Moving Average (SMA), Exponential Moving Average (EMA), Relative Strength Index (RSI), Moving Average Convergence Divergence (MACD), Daily Returns, and Volatility.


**Reasoning**:
The first step is to define the `calculate_sma` function as per the instructions, which will take a price Series and a window, then compute and return the Simple Moving Average.



````python
import pandas as pd
import numpy as np

def calculate_sma(prices, window):
    """
    Calculates the Simple Moving Average (SMA).

    Args:
        prices (pd.Series): A pandas Series of prices.
        window (int): The number of periods to consider for the SMA.

    Returns:
        pd.Series: A pandas Series containing the SMA values.
    """
    return prices.rolling(window=window).mean()
````

**Reasoning**:
The next step is to define the `calculate_ema` function as per the instructions, which will take a price Series and a window, then compute and return the Exponential Moving Average.



````python
def calculate_ema(prices, window):
    """
    Calculates the Exponential Moving Average (EMA).

    Args:
        prices (pd.Series): A pandas Series of prices.
        window (int): The number of periods to consider for the EMA.

    Returns:
        pd.Series: A pandas Series containing the EMA values.
    """
    return prices.ewm(span=window, adjust=False).mean()
````

**Reasoning**:
The next step is to define the `calculate_rsi` function, which will take a price Series and a window, compute the Relative Strength Index, and handle initial NaN values, then return it as a pandas Series.



````python
def calculate_rsi(prices, window):
    """
    Calculates the Relative Strength Index (RSI).

    Args:
        prices (pd.Series): A pandas Series of prices.
        window (int): The number of periods to consider for the RSI.

    Returns:
        pd.Series: A pandas Series containing the RSI values.
    """
    delta = prices.diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.ewm(span=window, adjust=False).mean()
    avg_loss = loss.ewm(span=window, adjust=False).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
````

## Define Financial Indicator Functions

### Subtask:
Define the `calculate_macd` function, which will compute the Moving Average Convergence Divergence (MACD) and its signal line based on short and long window EMAs, then return them as pandas Series.


**Reasoning**:
The next step is to define the `calculate_macd` function as per the instructions, which will compute the Moving Average Convergence Divergence (MACD) and its signal line based on short and long window EMAs, then return them as pandas Series.



````python
def calculate_macd(prices, short_window, long_window, signal_window):
    """
    Calculates the Moving Average Convergence Divergence (MACD) and its signal line.

    Args:
        prices (pd.Series): A pandas Series of prices.
        short_window (int): The short-period window for EMA.
        long_window (int): The long-period window for EMA.
        signal_window (int): The window for the signal line EMA.

    Returns:
        tuple: A tuple containing two pandas Series: (macd_line, signal_line).
    """
    exp1 = prices.ewm(span=short_window, adjust=False).mean()
    exp2 = prices.ewm(span=long_window, adjust=False).mean()
    macd_line = exp1 - exp2
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()
    return macd_line, signal_line
````

**Reasoning**:
The next step is to define the `calculate_daily_returns` function, which will compute the daily percentage change of prices and return it as a pandas Series.



````python
def calculate_daily_returns(prices):
    """
    Calculates the daily percentage returns.

    Args:
        prices (pd.Series): A pandas Series of prices.

    Returns:
        pd.Series: A pandas Series containing the daily percentage returns.
    """
    return prices.pct_change()
````

**Reasoning**:
The next step is to define the `calculate_volatility` function, which will compute the volatility (standard deviation of daily returns) over a specified window and return it as a pandas Series.



````python
def calculate_volatility(prices, window):
    """
    Calculates the volatility (standard deviation of daily returns).

    Args:
        prices (pd.Series): A pandas Series of prices.
        window (int): The number of periods to consider for volatility calculation.

    Returns:
        pd.Series: A pandas Series containing the volatility values.
    """
    daily_returns = prices.pct_change()
    volatility = daily_returns.rolling(window=window).std() * np.sqrt(252) # Annualized volatility
    return volatility
````

## Fetch Stock Data

### Subtask:
Obtain historical stock data for Reliance Industries Limited (RELIANCE.NS) for the past two years up to the most recent available date, as future data cannot be accessed.


**Reasoning**:
The first step is to import the `yfinance` library, which is necessary to fetch stock data as per the instructions.



````python
import yfinance as yf
import pandas as pd
import datetime

print("yfinance, pandas, and datetime libraries imported successfully.")
````

**Reasoning**:
As per the instructions, the next step is to define the ticker symbol for Reliance Industries Limited, calculate the start and end dates for the past two years, fetch the historical stock data using `yfinance`, and then extract the 'Close' prices into a pandas Series.



````python
ticker = 'RELIANCE.NS'
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=2*365) # Approximately two years

print(f"Fetching data for {ticker} from {start_date} to {end_date}")
data = yf.download(ticker, start=start_date, end=end_date)

reliance_prices = data['Close']

print("Stock data fetched and 'Close' prices extracted successfully.")
print(reliance_prices.head())
````

**Reasoning**:
Now that the historical stock data for Reliance Industries Limited has been fetched and the 'Close' prices extracted, the next step is to apply the financial indicator functions. As per the instructions, I will start by calculating the Simple Moving Average (SMA) using the `calculate_sma` function defined earlier.



````python
window_sma = 20
reliance_prices['SMA'] = calculate_sma(reliance_prices, window_sma)

print(f"SMA ({window_sma} days) calculated successfully.")
print(reliance_prices.tail())
````

**Reasoning**:
The previous step resulted in a `SettingWithCopyWarning` when adding the 'SMA' column because `reliance_prices` might have been a view. To ensure safe and independent modifications for further indicator calculations, `reliance_prices` is explicitly made a deep copy. Then, the next financial indicator, Exponential Moving Average (EMA), will be calculated using the `calculate_ema` function on the 'RELIANCE.NS' price column and added as a new column to the `reliance_prices` DataFrame.



````python
# Ensure reliance_prices is a standalone DataFrame to avoid SettingWithCopyWarning for subsequent calculations
reliance_prices = reliance_prices.copy(deep=True)

window_ema = 20
reliance_prices['EMA'] = calculate_ema(reliance_prices['RELIANCE.NS'], window_ema)

print(f"EMA ({window_ema} days) calculated successfully.")
print(reliance_prices.tail())
````

**Reasoning**:
Now that SMA and EMA have been calculated, the next step is to calculate the Relative Strength Index (RSI) using the `calculate_rsi` function. This will add the RSI values as a new column to the `reliance_prices` DataFrame.



````python
window_rsi = 14
reliance_prices['RSI'] = calculate_rsi(reliance_prices['RELIANCE.NS'], window_rsi)

print(f"RSI ({window_rsi} days) calculated successfully.")
print(reliance_prices.tail())
````

**Reasoning**:
Now that SMA, EMA, and RSI have been calculated, the next step is to calculate the Moving Average Convergence Divergence (MACD) and its signal line using the `calculate_macd` function. This will add both the MACD line and the Signal line as new columns to the `reliance_prices` DataFrame.



````python
short_window_macd = 12
long_window_macd = 26
signal_window_macd = 9

macd_line, signal_line = calculate_macd(
    reliance_prices['RELIANCE.NS'],
    short_window_macd,
    long_window_macd,
    signal_window_macd
)

reliance_prices['MACD'] = macd_line
reliance_prices['MACD_Signal'] = signal_line

print(f"MACD ({short_window_macd}, {long_window_macd}, {signal_window_macd}) calculated successfully.")
print(reliance_prices.tail())
````

**Reasoning**:
Now that SMA, EMA, RSI, and MACD have been calculated, the next step is to calculate the Daily Returns using the `calculate_daily_returns` function. This will add the daily percentage change in prices as a new column to the `reliance_prices` DataFrame.



````python
reliance_prices['Daily_Returns'] = calculate_daily_returns(reliance_prices['RELIANCE.NS'])

print("Daily Returns calculated successfully.")
print(reliance_prices.tail())
````

**Reasoning**:
Now that SMA, EMA, RSI, MACD, and Daily Returns have been calculated, the final financial indicator to be calculated is Volatility, using the `calculate_volatility` function. This will add the volatility values as a new column to the `reliance_prices` DataFrame, completing the calculation of all six required indicators.



````python
window_volatility = 20
reliance_prices['Volatility'] = calculate_volatility(reliance_prices['RELIANCE.NS'], window_volatility)

print(f"Volatility ({window_volatility} days) calculated successfully.")
print(reliance_prices.tail())
````
